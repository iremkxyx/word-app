```python
import matplotlib.pyplot as plt
import io
import base64

def create_pie_chart(correct, wrong):
    labels = ['Doƒüru', 'Yanlƒ±≈ü']
    sizes = [correct, wrong]
    colors = ['#4CAF50', '#F44336']  # ye≈üil, kƒ±rmƒ±zƒ±

    fig, ax = plt.subplots()
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
    ax.axis('equal')  # daire ≈üekli i√ßin

    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    plt.close(fig)
    buf.seek(0)
    img_bytes = buf.read()
    return img_bytes
```

```python
import gradio as gr
import json
import os
import random

USERS_FILE = "users.json"

def load_users():
    if not os.path.exists(USERS_FILE):
        return {}
    with open(USERS_FILE, "r") as f:
        try:
            return json.load(f)
        except json.JSONDecodeError:
            return {}

def save_users(users):
    with open(USERS_FILE, "w") as f:
        json.dump(users, f, indent=2)

def add_to_favorites(username, word):
    users = load_users()
    mesaj = ""
    favoriler = []
    if username in users:
        if "favori_kelimeler" not in users[username]:
            users[username]["favori_kelimeler"] = []
        if word not in users[username]["favori_kelimeler"]:
            users[username]["favori_kelimeler"].append(word)
            save_users(users)
            mesaj = "‚úÖ Favorilere eklendi"
        else:
            mesaj = "Zaten favorilerde"
        favoriler = users[username]["favori_kelimeler"]
    else:
        mesaj = "Kullanƒ±cƒ± bulunamadƒ±"
    

    favori_str = ", ".join(favoriler)

    return f"{favori_str}\n{mesaj}"


def get_favorites(username):
    users = load_users()
    return users.get(username, {}).get("favori_kelimeler", [])

def remove_favorite(username, word):
    users = load_users()
    if username in users and word in users[username].get("favori_kelimeler", []):
        users[username]["favori_kelimeler"].remove(word)
        save_users(users)
    return get_favorites(username)
def show_login_form(mode):
    return (
        gr.update(visible=True),  # username_input
        gr.update(visible=True),  # password_input
        gr.update(visible=True),  # onay_btn
        mode,                     # login_mode
        gr.update(visible=False)  # login_output (gizle)
    )
    
def handle_auth(username, password, mode):
    users = load_users()
    if mode == "login":
        if username in users and users[username]["password"] == password:
            user_data = users[username]
            correct = user_data.get("correct", 0)
            wrong = user_data.get("wrong", 0)
            daily_goal = user_data.get("daily_goal", 10)
            return (
                gr.update(value=f"üéâ Ho≈ü geldin, {username}!", visible=True),
                username,
                correct,
                wrong,
                daily_goal,
                gr.update(visible=False),
                gr.update(visible=False),
                gr.update(visible=False),
                gr.update(visible=True),
                gr.update(visible=True),
                gr.update(visible=True),
                gr.update(visible=True)
            )
        else:
            return (
                gr.Textbox.update(value="‚ùå Hatalƒ± kullanƒ±cƒ± adƒ± veya ≈üifre", visible=True),
                "", 0, 0, 10,
                gr.update(visible=True),
                gr.update(visible=True),
                gr.update(visible=True),
                gr.update(visible=False),
                gr.update(visible=False),
                gr.update(visible=False),
                gr.update(visible=False)
            )
    elif mode == "register":
        if username in users:
            return (
                gr.update(value="‚ùå Bu kullanƒ±cƒ± adƒ± zaten alƒ±nmƒ±≈ü", visible=True),
                "", 0, 0, 10,
                gr.update(visible=True),
                gr.update(visible=True),
                gr.update(visible=True),
                gr.update(visible=False),
                gr.update(visible=False),
                gr.update(visible=False),
                gr.update(visible=False)
            )
        users[username] = {
            "password": password,
            "correct": 0,
            "wrong": 0,
            "daily_goal": 10,
            "zayif_kelimeler": {},
            "favori_kelimeler": []
        }
        save_users(users)
        return (
            gr.update(value=f"üéâ Kayƒ±t ba≈üarƒ±lƒ±, ho≈ü geldin {username}!", visible=True),
            username,
            0,
            0,
            10,
            gr.update(visible=False),
            gr.update(visible=False),
            gr.update(visible=False),
            gr.update(visible=True),
            gr.update(visible=True),
            gr.update(visible=True),
            gr.update(visible=True)
        )
        
def get_rank(correct):
    return "üî∞ √áaylak" if correct < 20 else "üõ†Ô∏è √áƒ±rak" if correct < 60 else "üéì Tecr√ºbeli" if correct < 100 else "üèÜ Usta"


def test(word, answer):
    correct = words.get(word, "")
    result = "‚úÖ Doƒüru" if answer.lower().strip() == correct else "‚ùå Yanlƒ±≈ü"
    new_word = random.choice(list(words))
    return result, correct, new_word

def check_answer(word, answer, correct, wrong, goal, username):
    result, correct_ans, new_word = test(word, answer)
    if result.startswith("‚úÖ"):
        correct += 1
    else:
        wrong += 1
    done = f"Tebrikler! {goal} kelime tamamlandƒ±." if correct >= goal else f"{goal - correct} kelime kaldƒ±."

    if username:
        users = load_users()
        u = users.get(username, {"correct": 0, "wrong": 0, "daily_goal": goal, "zayif_kelimeler": {}})
        u.setdefault("zayif_kelimeler", {})
        u["correct"], u["wrong"] = correct, wrong
        if result.startswith("‚ùå"):
            u["zayif_kelimeler"][word] = u["zayif_kelimeler"].get(word, 0) + 1
        users[username] = u
        save_users(users)

    return (
        result,
        correct_ans,
        new_word,                # Yeni kelimeyi word_display g√ºncellemek i√ßin g√∂nderiyoruz
        gr.update(value=done, visible=True),
        correct,
        wrong,
        None,                   # audio_output (≈üimdilik bo≈ü)
        gr.update(value=sentences[new_word], visible=True),  # Yeni kelimenin c√ºmlesi g√∂sterilecek
        gr.update(value=correct_ans, visible=True)
    )

def update_stats(correct, wrong, goal):
    total = correct + wrong
    cr = int(correct / total * 100) if total else 0
    wr = int(wrong / total * 100) if total else 0
    prog = int(correct / goal * 100) if goal else 0
    return f"%{prog} tamamlandƒ±", f"Toplam: {total}", f"Doƒüru: %{cr}", f"Yanlƒ±≈ü: %{wr}", get_rank(correct)

def get_weak_words(username):
    u = load_users().get(username, {}).get("zayif_kelimeler", {})
    return [w for w, _ in sorted(u.items(), key=lambda x: x[1], reverse=True)[:5]]

def show_weak(username):
    lst = get_weak_words(username)
    return ", ".join(lst), lst[0] if lst else ""




    
with gr.Blocks() as demo:

    user_state = gr.State("")
    correct_state = gr.State(0)
    wrong_state = gr.State(0)
    goal_state = gr.State(10)

    with gr.Tabs():
        with gr.TabItem("Giri≈ü/Kayƒ±t"):
            gr.Markdown("## üë§ Giri≈ü veya Kayƒ±t Ol")
            with gr.Row():
                login_btn = gr.Button("Giri≈ü Yap")
                register_btn = gr.Button("Kayƒ±t Ol")
            username_input = gr.Textbox(label="Kullanƒ±cƒ± Adƒ±", visible=False)
            password_input = gr.Textbox(label="≈ûifre", type="password", visible=False)
            onay_btn = gr.Button("Onayla", visible=False)
            login_mode = gr.Textbox(visible=False)
            login_output = gr.Textbox(label="Durum", visible=False)

     



    
        with gr.TabItem("Kelime √ñƒürenme", visible=False) as kelime_tab:
            gr.Markdown("## üìò Kelime √ñƒürenme")
            word_display = gr.Textbox(label="Kelime", value=random.choice(list(words.keys())), interactive=False)
            fav_output = gr.Textbox(label="Durum", interactive=False)
            example_sentence_box = gr.Textbox(label="üìö √ñrnek C√ºmle", interactive=False, visible=True, value="")
            answer_input = gr.Textbox(label="T√ºrk√ße anlamƒ±nƒ± yaz")
            result_output = gr.Textbox(label="Sonu√ß", interactive=False)
            correct_answer_output = gr.Textbox(label="Doƒüru Cevap", visible=False, interactive=False)
            done_box = gr.Textbox(label="Mesaj", interactive=False, visible=False)
            audio_output = gr.Audio(label="Ses", type="filepath")
            submit_btn = gr.Button("Cevabƒ± G√∂nder")
            fav_btn = gr.Button("‚≠ê Favorilere Ekle")  
        
            submit_btn.click(
                fn=check_answer,
                inputs=[word_display, answer_input, correct_state, wrong_state, goal_state, user_state],
                outputs=[
                    result_output,
                    correct_answer_output,
                    word_display,       # Burada word_display g√ºncelleniyor
                    done_box,
                    correct_state,
                    wrong_state,
                    audio_output,
                    example_sentence_box,
                    correct_answer_output
                ]
            )
        fav_btn.click(fn=add_to_favorites, inputs=[user_state, word_display], outputs=fav_output)


        with gr.TabItem("ƒ∞statistikler", visible=False) as istatistikler_tab:
            gr.Markdown("## üìä ƒ∞statistikler")
            update_btn = gr.Button("ƒ∞statistikleri G√ºncelle")
            progress = gr.Label()
            total = gr.Label()
            ratio_c = gr.Label()
            ratio_w = gr.Label()
            rank = gr.Label()
            update_btn.click(update_stats, [correct_state, wrong_state, goal_state], [progress, total, ratio_c, ratio_w, rank])

        with gr.TabItem("Tekrar Listem", visible=False) as tekrar_tab:
            gr.Markdown("## ‚ôªÔ∏è Zayƒ±f Olduƒüun Kelimeler")
            weak_words_box = gr.Textbox(label="Kelime Listesi", interactive=False)
            weak_show_btn = gr.Button("G√∂ster")
            weak_word_display = gr.Textbox(value="", label="Zayƒ±f Kelime", interactive=False)
            weak_answer_input = gr.Textbox(label="Anlamƒ±")
            weak_result_output = gr.Textbox(label="Sonu√ß")
            weak_show_btn.click(fn=show_weak, inputs=[user_state], outputs=[weak_words_box, weak_word_display])
            weak_submit_btn = gr.Button("Zayƒ±f Kelimeyi Yanƒ±tla")
            weak_submit_btn.click(
                check_answer,
                [weak_word_display, weak_answer_input, correct_state, wrong_state, goal_state, user_state],
                [weak_result_output, correct_answer_output, weak_word_display, done_box, correct_state, wrong_state, audio_output, example_sentence_box, correct_answer_output]
             )
        with gr.TabItem("Favori Kelimelerim", visible=False) as favori_tab:
            gr.Markdown("## ‚≠ê Favorilere Eklenen Kelimeler")
            favorites_box = gr.Textbox(label="Favoriler", interactive=False)
            refresh_btn = gr.Button("Listeyi Yenile")
            remove_input = gr.Textbox(label="Silmek ƒ∞stediƒüin Kelime")
            remove_btn = gr.Button("Kelimeleri Sil")

            refresh_btn.click(fn=get_favorites, inputs=[user_state], outputs=favorites_box)
            remove_btn.click(fn=remove_favorite, inputs=[user_state, remove_input], outputs=favorites_box)

    login_btn.click(fn=lambda: show_login_form("login"), inputs=[], outputs=[username_input, password_input, onay_btn, login_mode, login_output])
    register_btn.click(fn=lambda: show_login_form("register"), inputs=[], outputs=[username_input, password_input, onay_btn, login_mode, login_output])
    onay_btn.click(
        handle_auth,
        [username_input, password_input, login_mode],
        [
            login_output, user_state, correct_state, wrong_state, goal_state,
            username_input, password_input, onay_btn,
            kelime_tab, istatistikler_tab, tekrar_tab, favori_tab  # t√ºm sekmeleri g√∂r√ºn√ºr yapƒ±yoruz
        ]
    )

    def after_login(*args):
        return (*args, gr.update(visible=False), gr.update(visible=False), gr.update(visible=False), gr.update(visible=True), gr.update(visible=True), gr.update(visible=True))

   



demo.launch()
```
